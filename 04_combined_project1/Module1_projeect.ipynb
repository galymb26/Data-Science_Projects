```python
# Создаем сводную таблицу для подсчета количества игр по годам выпуска
years_data = pd.pivot_table(games_data, index='year_of_release', values='name', aggfunc='count')

# Переименовываем столбец
years_data.columns = ['games']

years_data
```

**3.2 Посмотрите, как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?**

# Шаг 5. Проверьте гипотезы

# Гипотеза 1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

**Нулевая и альтернативная гипотезы:**

**Гипотеза Н0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**

**Гипотеза Н1: средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.**

<font color='DarkBlue'><b>Комментарий ревьюера</b></font><br>
<font color='DarkGreen'>👌 Хорошо, правильно сформулированы гипотезы.</font>

```python
# Отфильтруем данные по платформам и удаление пропусков в пользовательских рейтингах
xbox_one_ratings = current_games_data[(current_games_data['platform'] == 'XOne') & (~current_games_data['user_score'].isna())]['user_score']
pc_ratings = current_games_data[(current_games_data['platform'] == 'PC') & (~current_games_data['user_score'].isna())]['user_score']

# Выполняем t-тест для независимых выборок
results = st.ttest_ind(xbox_one_ratings, pc_ratings, equal_var=False)  

print(results.pvalue)

# Проверяем статистическую значимость
alpha = 0.05
if results.pvalue < 0.05:
    print("Отвергаем нулевую гипотезу: средние пользовательские рейтинги Xbox One и PC не равны.")
else:
    print("Не отвергаем нулевую гипотезу: нет статистически значимой разницы между средними пользовательскими рейтингами Xbox One и PC.")
```

**Вывод:**

**Не отвергаем нулевую гипотезу.** 

**Таким образом, мы можем принять утверждение, что средние пользовательские рейтинги платформ Xbox One и PC не отличаются. Это может быть связано с тем что пользователи Xbox One и PC часто могут быть одной и той же аудиторией, так как Microsoft активно развивает экосистему, которая позволяет игры на Xbox быть доступными и на PC через Windows Store или Xbox Game Pass для PC. Это ведет к тому, что пользователи обеих платформ имеют доступ к одним и тем же играм, что может приводить к похожим оценкам.**

**Нулевая гипотеза (H0):
Нулевая гипотеза формулируется с предположением о том, что нет никакого эффекта или различия между группами или образцами данных. В этом случае, нулевая гипотеза звучит так "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые". Это предполагает, что нет статистически значимого различия в средних пользовательских рейтингах между платформами Xbox One и PC.**

**Альтернативная гипотеза (H1):
Альтернативная гипотеза выражает противоположное предположение: что между группами или образцами данных есть какое-то различие или эффект. В вашем случае, альтернативная гипотеза звучит как "Средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой". Это предполагает, что существует статистически значимое различие в средних пользовательских рейтингах между платформами Xbox One и PC.**

# Гипотеза 2: Средние пользовательские рейтинги жанров Action  и Sports разные.

**Нулевая и альтернативная гипотезы:**

**Гипотеза Н0: средние пользовательские рейтинги жанров Action и Sports одинаковые.**

**Гипотеза Н1: средние пользовательские рейтинги жанров Action и Sports отличаются между собой**

```python
# Отфильтруем данные по жанрам и удаление пропусков в пользовательских рейтингах
action_ratings = current_games_data[(current_games_data['genre'] == 'Action') & (~current_games_data['user_score'].isna())]['user_score']
sports_ratings = current_games_data[(current_games_data['genre'] == 'Sports') & (~current_games_data['user_score'].isna())]['user_score']

# Рассчитываем средние пользовательские рейтинги для жанров
action_mean_rating = action_ratings.mean()
sports_mean_rating = sports_ratings.mean()

# Выполняем t-тест для независимых выборок
results = st.ttest_ind(action_ratings, sports_ratings, equal_var=False)

print("Средний пользовательский рейтинг для жанра Action:", action_mean_rating)
print("Средний пользовательский рейтинг для жанра Sports:", sports_mean_rating)
print("")

print(results.pvalue)

# Проверяем статистическую значимость
alpha = 0.05
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sports различаются.")
else:
    print("Не отвергаем нулевую гипотезу: нет статистически значимой разницы между средними пользовательскими рейтингами жанров Action и Sports.")
```

**Вывод:**

**Отвергаем нулевую гипотезу. Таким образом, мы можем принять утверждение, что средние пользовательские рейтинги жанров Action и Sports отличаются.** 

**Нулевая гипотеза (H0) обычно формулируется так, чтобы предположить отсутствие эффекта или различия между группами. В нашем случае, мы формулируем нулевую гипотезу как "средние пользовательские рейтинги жанров Action и Sports одинаковые". Это предположение означает, что нет статистически значимого различия в средних пользовательских рейтингах между этими двумя жанрами.**

**Альтернативная гипотеза (H1) предполагает, что существует какое-то различие между группами. В нашем случае, альтернативная гипотеза сформулирована как "средние пользовательские рейтинги жанров Action и Sports отличаются между собой". Это предположение означает, что есть статистически значимое различие в средних пользовательских рейтингах между жанрами Action и Sports.**
